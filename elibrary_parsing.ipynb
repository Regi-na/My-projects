```python
soup = BeautifulSoup(browser.page_source, 'lxml')
list__ = soup.find_all(title='Информация о журнале')
```

```python
button2 = browser.find_element_by_xpath('//*[@id="pages"]/table/tbody/tr/td[4]/a')
button2.click()
```

```python
soup2 = BeautifulSoup(browser.page_source, 'lxml')
list__2 = soup2.find_all(title='Информация о журнале')
```

```python
id = []
names = []
for link in list__:
  id.append(re.findall("[-+]?[.]?[\d]+(?:,\d\d\d)*[\.]?\d*(?:[eE][-+]?\d+)?",(link.attrs['href'])))
  names.append(link.get_text())

for link in list__2:
  id.append(re.findall("[-+]?[.]?[\d]+(?:,\d\d\d)*[\.]?\d*(?:[eE][-+]?\d+)?",(link.attrs['href'])))
  names.append(link.get_text())
```

```python
numbs = []
for i in (id):
  for line in i:
    if line.strip():           # line contains eol character(s)
       numbs.append(int(line))
```

```python
df = pd.DataFrame({'id': numbs, 'name': names})
```

```python
df
```

```python
for i in range(len(numbs)):
  d = dict.fromkeys(["Список статей, опубликованных в журнале в {} году".format(k) for k in range(2011, 2021)],0)
  link = str('https://www.elibrary.ru/title_profile.asp?id=' + str(numbs[i]))
  browser  = webdriver.Chrome('chromedriver', chrome_options=chrome_options)
  browser.get(link)

  soup = BeautifulSoup(browser.page_source, 'lxml')
  for c in range(2011, 2021):
    list__ = soup.findAll(title="Список статей, опубликованных в журнале в {} году".format(c))
    for link in list__:
      d.update({link.attrs['title']: int(link.get_text())})

  df.loc[df.id == numbs[i], list(d.keys())] = list(d.values())
```

```python
name = ['id', 'name']
for i in range (2011,2021):
  name.append('n_papers_risc_{}'.format(i))

df.set_axis(name, axis='columns', inplace=True)
```
